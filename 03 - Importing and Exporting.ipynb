## Deletion of Missing Values

```python
#how="any" & axis=0 {means that if Python finds any missing value in any row, that particular row will get deleted}
pla.dropna(how="any",axis=0).shape
```

```python
#how = "all" & axis=0 {means that Python will only delete the row if and only if that particular row has all the missing values }
pla.dropna(how="all",axis=0).shape
```

```python
#how="any" & axis=1 {means that python will delete the whole column if there is any missing value in that particular column}
pla.dropna(how="any",axis=1).shape
```

```python
#how ="all" & axis=1 {means that python will only delete the whole colum if that column have all the missing value.}
pla.dropna(how="all",axis=1).shape
```
